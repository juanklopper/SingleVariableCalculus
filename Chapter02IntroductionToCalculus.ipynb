## 2.1 Defining calculus

Calculus is the branch of mathematics that deals with change. It is used to study how quantities change over time, and to solve problems that involve rates of change, such as the speed of a moving object or the acceleration of a falling body. Calculus is also used to find the areas and volumes of shapes, and to calculate the forces that act on objects.

Calculus allows us to describe natural phenomena using mathematical equations. It is a tool that can be used to solve a wide variety of problems in healthcare.

Calculus is divided into two main branches: differential calculus and integral calculus.

* Differential calculus is concerned with rates of change.
* Integral calculus is concerned with the accumulation of quantities over time.

```python
# Create an array of 200 values on the interval from 55 to 85 assigned to the variable age
age = np.linspace(55, 85, 200)
```

```python
# Create an array that multiplies the age array by 1.5 and adds 10 and assigns it to the variable dbp
dbp = 1.5 * age + 10
```

```python
# Create a line plot of the age array on the x-axis and the dbp array on the y-axis and add a grid and a title named "Diastolic Blood Pressure"
# Add a label to the x-axis named "Age (years)" and a label to the y-axis named "Diastolic Blood Pressure (mmHg)"
# Remove all spines from the plot
plt.plot(age, dbp)
plt.grid()
plt.title("Diastolic Blood Pressure")
plt.xlabel("Age (years)")
plt.ylabel("Diastolic Blood Pressure (mmHg)")
plt.box(False)
plt.show();
```

```python
# Recreate the previous plot and add a line from point (60,100) to (70,100)
# Add a line from point (70,100) to (70,115)

plt.plot(age, dbp)
plt.grid()
plt.title("Diastolic Blood Pressure")
plt.xlabel("Age (years)")
plt.ylabel("Diastolic Blood Pressure (mmHg)")
plt.box(False)
plt.plot([60, 70], [100, 100], color='red')
plt.plot([70, 70], [100, 115], color='red')
plt.show();
```

```python
# Generate a user-defined function that takes an array of values as imput and calculate the probability density of each value using the normal distribution
# Assign the function name to the variable name pdf
def pdf(x):
    return 1 / np.sqrt(2 * np.pi) * np.exp(-x ** 2 / 2)
```

```python
# Create an array of 300 values on the interval from -4 to 4 assigned to the variable x
x = np.linspace(-4, 4, 300)
```

```python
# Create a plot of the x array on the x-axis and the pdf function on the y-axis
# Add a grid and a title named "Standard Normal Distribution"
# Add a label to the x-axis named "x" and a label to the y-axis named "Probability Density"
# Remove all spines from the plot
plt.plot(x, pdf(x))
plt.grid()
plt.title("Standard Normal Distribution")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.box(False)
plt.show();
```

```python
# Recreate the plot above and add red line from point (-1.96,0) to (-1.96,0.058)

plt.plot([-1.96, -1.96], [0, 0.058], color='red')
plt.plot(x, pdf(x))
plt.grid()
plt.title("Standard Normal Distribution")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.box(False)
plt.show();
```

## 2.3 Math of the microscopic

Calculus is based on the idea of the infinitesimal. An infinitesimal is a quantity that is so small that it cannot be measured.

```python
# Create an array of 400 values on the interval from 5 to 6 assigned to the variable x
x = np.linspace(5, 6, 400)
```

```python
# Create a line plot of the array x on the x-axis and the function sin(x**4)/(x**2) on the y-axis
# Add a grid and a title named "A Polynomial Function"
# Add a label to the x-axis named "x" and a label to the y-axis named "y"
# Remove all spines from the plot

plt.plot(x, np.sin(x**4)/(x**1))
plt.grid()
plt.title("A Polynomial Function")
plt.xlabel("x")
plt.ylabel("y")
plt.box(False)
plt.show();
```
